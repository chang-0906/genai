```python
import os
from langfuse.openai import OpenAI

client = OpenAI(
    base_url="https://chang-0906--ollama-gguf-hf-import-ollamaggufserver-api.modal.run/v1",
    api_key="ollama",
)

messages = [
    {"role": "system", "content": "請一律用繁體中文回答。"},
]

def chat(user_input):
    messages.append({"role": "user", "content": user_input})

    resp = client.chat.completions.create(
        name="notebook-chat",
        model="llama31-abliterated-q4km:latest",
        messages=messages,
        metadata={
            "app": "modal-notebook",
            "provider": "modal-ollama",
        },
        extra_body={
            "langfuse_user_id": "notebook-user",
            "langfuse_session_id": "notebook-session",
            "langfuse_tags": ["notebook", "modal", "ollama"],
        },
    )

    reply = resp.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    return reply
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML

input_box = widgets.Text(
    placeholder="輸入訊息...",
    layout=widgets.Layout(width='80%')
)

output_area = widgets.Output()

display(output_area, input_box)

def handle_submit(change):
    user_input = change["new"]
    if not user_input:
        return

    input_box.value = ""

    messages.append({"role": "user", "content": user_input})

    resp = client.chat.completions.create(
        model="llama31-abliterated-q4km:latest",
        messages=messages,
    )

    reply = resp.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})

    with output_area:
        display(HTML(f"""
        <div style="text-align:right; margin:10px">
            <b>你:</b> {user_input}
        </div>
        <div style="text-align:left; margin:10px">
            <b>AI:</b> {reply}
        </div>
        """))

input_box.observe(handle_submit, names='value')
```
